```python
import warnings
with warnings.catch_warnings():
    warnings.filterwarnings("ignore",category=FutureWarning)
    import tensorflow as tf

import numpy as np
import pandas as pd
import itertools
import os
```

```python
predictStd = 201630
```

```python
df = pd.read_csv('../dataset/feature_regression_example.csv')
```

```python
df = df.sort_values(by='YEARWEEK')
```

```python
df.head()
```

```python
len(df)
```

```python
df.dtypes
```

```python
train_set = df.query('YEARWEEK <= @predictStd')
```

```python
train_set.head()
```

```python
len(train_set)
```

```python
test_set = df.query('YEARWEEK > @predictStd')
```

```python
test_set = test_set.reset_index(drop=True)
```

```python
test_set.head()
```

```python
len(test_set)
```

```python
predict_set = test_set
predict_set['QTY_FCST'] = np.nan
```

```python
predict_set.head()
```

```python
features_key = ['WEEK', 'HCLUS', 'PRO_PERCENT']
label_key = 'QTY'
```

```python
feature_cols = [tf.feature_column.numeric_column(k) for k in features_key]
```

```python
feature_cols
```

```python
estimator = tf.estimator.LinearRegressor(feature_columns=feature_cols, model_dir="train")
```

```python
def get_input_fn(data_set, num_epochs=None, n_batch = 128, shuffle=True):
         return tf.estimator.inputs.pandas_input_fn(
         x = pd.DataFrame({k: data_set[k].values for k in features_key}),
         y = pd.Series(data_set[label_key].values),
         batch_size=n_batch,
         num_epochs=num_epochs,
         shuffle=shuffle)
```

```python
estimator.train(input_fn=get_input_fn(train_set, num_epochs=None, n_batch = 128, shuffle=False), steps=1000)
```

```python
ev = estimator.evaluate(    
          input_fn=get_input_fn(test_set.sort_values(by='YEARWEEK'),                          
          num_epochs=1,                          
          n_batch = 128,                          
          shuffle=False))
```

```python
loss_score = ev["loss"]
print("Loss: {0:f}".format(loss_score))
```

```python
train_set['QTY'].describe()
```

```python
len(predict_set)
```

```python
y = estimator.predict(    
         input_fn=get_input_fn(predict_set.sort_values(by='YEARWEEK'),                          
         num_epochs=1,                          
         n_batch = 128,                          
         shuffle=False))
```

```python
predictions = list(p["predictions"] for p in itertools.islice(y, 22))
print("Predictions: {}".format(str(predictions)))
```

```python
pred = pd.DataFrame(predictions)
```
